```python
TMB_crew = pd.read_csv(Dataset_directory+"Processed"+os.sep+"TMB_crew"+".csv",index_col="Unnamed: 0")
TMB_crew.index = TMB_crew["credit_id"]

TMB_crew_individuals = pd.read_csv(Dataset_directory+"Processed"+os.sep+"TMB_crew_individuals"+".csv",index_col="Unnamed: 0")
TMB_crew_individuals.index = TMB_crew_individuals["id"]

TMB_cast = pd.read_csv(Dataset_directory+"Processed"+os.sep+"TMB_cast"+".csv",index_col="Unnamed: 0")
TMB_cast.index = TMB_cast["credit_id"]

TMB_cast_individuals = pd.read_csv(Dataset_directory+"Processed"+os.sep+"TMB_cast_individuals"+".csv",index_col="Unnamed: 0")
TMB_cast_individuals.index = TMB_cast_individuals["id"]

# 98% accuracy (assuming TMB gender is correct!) for "male"/"female" (84% of the names)
# 78% accuract for "mostly_female/mostly_male" (7% of the names)
# unknown -> 8%
# androgeneous -> 1.5%

# !pip install gender-guesser

import gender_guesser.detector as gender
d = gender.Detector(case_sensitive=False)
TMB_crew_individuals["name"] = TMB_crew_individuals["name"].apply(lambda x: x.strip())
TMB_cast_individuals["name"] = TMB_cast_individuals["name"].apply(lambda x: x.strip())

TMB_crew_individuals["gender_guesser"] = TMB_crew_individuals["name"].apply(lambda x: d.get_gender( x.split(" ")[0] ) )
TMB_cast_individuals["gender_guesser"] = TMB_cast_individuals["name"].apply(lambda x: d.get_gender( x.split(" ")[0] ) )



importulence = TMB_cast_individuals["credit_number"]/TMB_cast_individuals["mean_order"]
TMB_cast_individuals["importulence"] = importulence.replace([np.inf, -np.inf], np.nan)
TMB_cast_individuals.sort_values(by = "importulence", ascending=False).head(200)

importulence = TMB_crew_individuals["credit_number"]/TMB_crew_individuals["mean_order"]
TMB_crew_individuals["importulence"] = importulence.replace([np.inf, -np.inf], np.nan)
TMB_crew_individuals.sort_values(by = "importulence", ascending=False).head(200)



TMB_cast_individuals["movieId_list"] = TMB_cast_individuals["credit_list"].apply(lambda x: eval(x.replace(" ",","))).apply(lambda x: [TMB_cast.loc[y, "movieId"] for y in x])
TMB_crew_individuals["movieId_list"] = TMB_crew_individuals["credit_list"].apply(lambda x: eval(x.replace(" ",","))).apply(lambda x: [TMB_crew.loc[y, "movieId"] for y in x])
```

```python
# if you're wondering why only two genders it's because this sort of approach lacks the resolution to even be *that confident*
# detecting people who do identify as one of two genders for their whole life

# NOTE: Could use intermediate values to rep uncertainty
# ... seems more dodgy than not unless you can quantify what the numbers *ought* to be
# which we could 
def gender_pidgeonholer(x):
    if x in ["female","mostly_female"]:
        return 1
    elif x in ["male","mostly_male"]:
        return 2
    else: # default don't know
        return 0
```

```python
gender_unknown = TMB_cast_individuals["gender"] == 0
TMB_cast_individuals["gender_guess_integrated"] = TMB_cast_individuals["gender"]
TMB_cast_individuals.loc[gender_unknown,"gender_guess_integrated"] = TMB_cast_individuals["gender_guesser"][gender_unknown].apply(gender_pidgeonholer)
TMB_cast_individuals["gender_guess_integrated"] = TMB_cast_individuals["gender_guess_integrated"]-1
TMB_cast_individuals["gender_guess_integrated"] = TMB_cast_individuals["gender_guess_integrated"].replace(-1, np.nan)
TMB_cast_individuals["gender_guess_integrated"].value_counts(dropna=False)
# 0=Female, 1=Male, unknown = NaN
```

```python
import shutil
import tempfile
import urllib.request
import gzip
import datetime

Dataset_directory = ".."+os.sep+"Datasets"+os.sep
tmdb_dir = create_subdir(Dataset_directory, "themoviedb")
```

```python

utc_now = datetime.datetime.utcnow()

if utc_now.hour<8: # upload for 'today' not necessary donw yet
    utc_now = today - datetime.timedelta(days = 1)
    
daily_file_export_list = ["movie","tv_series","person","collection","tv_network","keyword","production_company"]

for daily_export in daily_file_export_list:

    daily_file_export_file = "http://files.tmdb.org/p/exports/"+daily_export+"_ids_"+utc_now.strftime("%m_%d_%Y")+".json.gz"
    # movie_ids_file

    outFilePath = tmdb_dir + daily_file_export_file.split("/")[-1].replace(".gz","")

    with urllib.request.urlopen(daily_file_export_file) as response:
        with tempfile.NamedTemporaryFile(delete=False) as tmp_file:
            shutil.copyfileobj(response, tmp_file)

    with gzip.open(tmp_file.name, mode='rt', encoding="utf-8") as f_in:
        with open(outFilePath, 'w', encoding="utf-8") as f_out:
            shutil.copyfileobj(f_in, f_out)

    daily_file_export_ids = pd.read_json(outFilePath, lines=True)
    daily_file_export_ids.describe(include='all')        
```

```python
file_path = tmdb_dir + "collections.msgpack"

collections = pd.read_msgpack(file_path)

collections.replace("",np.nan,inplace=True)
collections.replace("No overview found.",np.nan,inplace=True)
collections.replace("No overview",np.nan,inplace=True)

collections["movie_ids"] = collections["parts"].apply(lambda x: [y["id"] for y in x])
collections["no_movies_in_collection"] = collections["movie_ids"].apply(lambda x: len(x))
collections.drop("parts",axis=1,inplace=True)
# I'm dropping collections with 0/1 movie id - they look like junk
collections.drop(collections[ collections["no_movies_in_collection"]==0 ].index, inplace=True)
collections.drop(collections[ collections["no_movies_in_collection"]==1 ].index, inplace=True)

collections.to_msgpack(tmdb_dir + "collections_cleaned.msgpack")
```
