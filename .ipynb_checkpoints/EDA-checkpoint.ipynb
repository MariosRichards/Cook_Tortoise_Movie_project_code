```python
import seaborn as sns
import matplotlib.pyplot as plt
sns.set(style="whitegrid")

def horiz_bar_chart_with_N(df,var1,var2,mask,min_sample_size,x_label,log):

    # Initialize the matplotlib figure
    f, ax = plt.subplots(figsize=(15, 6))

#     df = INT_df[[ var2, var1]][mask]
#     df[var1] = df[var1].apply(lambda x: eval(x)[0])
    df_N = df.groupby(var1).count().reset_index()
    df = df.groupby(var1).mean().reset_index()
    df["N"] = df_N[var2]
    decent_sample_size_indices = df_N[df_N[var2]>=min_sample_size].index
    df = df.loc[decent_sample_size_indices].sort_values(by=var2,ascending=True)
    if tail_head:
        if tail_head<0:
            df = df.tail(-tail_head)
        else:
            df = df.head(tail_head)
        

    sns.set_color_codes("pastel")
    g = sns.barplot(x=var2, y=var1, data=df,
                label=x_label, color="b");


    # Add a legend and informative axis label
    # ax.legend(ncol=2, loc="lower right", frameon=True)
    # ax.set(xlim=(0, df[var2].max()), ylabel="",
    #        xlabel="Gross/Budget")

    sns.despine(left=True, bottom=True)
    if log == True:

        g.set_xscale('log')

    df = df.reset_index().drop("index",axis=1)
    count=0
    for index, row in df.iterrows():
        g.text(row[var2], row.name, 'N=%d' % row["N"], fontsize=10)
        count=count+1

    plt.show()
    return df
```

```python
var2 = "gross_by_budget"
var1 = "TMB_production_countries"
mask = ( INT_df[var2].notnull() ) & (INT_df["TMB_num_production_countries"]==1.0)
min_sample_size = 20
x_label = "Gross/Budget"
log = True
tail_head = False

df = INT_df[[ var2, var1]][mask]
df[var1] = df[var1].apply(lambda x: eval(x)[0])

horiz_bar_chart_with_N(df,var1,var2,mask,min_sample_size,x_label,log)
```
